### Analyzing NYC Public School SAT Performance

We will address the following questions regarding the SAT performance of NYC public schools:

1. **Finding schools with the best math scores**
2. **Identifying the top 10 performing schools**
3. **Locating the NYC borough with the largest standard deviation in SAT performance**

Let's start by analyzing the dataset to find the schools with the best math scores.

```python
# Import necessary library
import pandas as pd

# Sample data for demonstration purposes
data = {
    'School Name': ['School A', 'School B', 'School C', 'School D'],
    'Math Score': [88, 92, 79, 85]
}

# Create a DataFrame
schools = pd.DataFrame(data)

# Finding schools with the best math scores
# Sort the schools by the Math score in descending order and display the top entries
best_math_schools = schools.sort_values(by='Math Score', ascending=False)
best_math_schools[['School Name', 'Math Score']].head()
```

```python
import pandas as pd

# Load your dataset
schools = pd.read_csv('schools.csv')

# Clean column names
schools.columns = schools.columns.str.strip().str.lower()

# Rename to match expected column names
schools = schools.rename(columns={
    'average_math': 'Math Score',
    'average_reading': 'Reading Score',
    'average_writing': 'Writing Score',
    'borough': 'Borough'
})

# Ensure required columns are present
required_columns = ['school_name', 'Math Score', 'Reading Score', 'Writing Score', 'Borough']
missing = [col for col in required_columns if col not in schools.columns]
if missing:
    raise KeyError(f"❌ Missing expected columns after rename: {missing}")

# Convert SAT scores to numeric just in case
for col in ['Math Score', 'Reading Score', 'Writing Score']:
    schools[col] = pd.to_numeric(schools[col], errors='coerce')

# === Task 1: Best Math Schools ===
schools['average_math'] = schools['Math Score']
best_math_schools = schools[schools['average_math'] >= 640][['school_name', 'average_math']]
best_math_schools = best_math_schools.sort_values(by='average_math', ascending=False).head(10)

# === Task 2: Top 10 by Total SAT ===
schools['total_SAT'] = (
    schools['Math Score'] +
    schools['Reading Score'] +
    schools['Writing Score']
)
top_10_schools = schools[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False).head(10)

# === Task 3: Borough with Largest Std Dev in Total SAT ===
borough_stats = schools.groupby('Borough').agg(
    num_schools=('school_name', 'count'),
    average_SAT=('total_SAT', 'mean'),
    std_SAT=('total_SAT', 'std')
).reset_index()

# Round to two decimal places
borough_stats['average_SAT'] = borough_stats['average_SAT'].round(2)
borough_stats['std_SAT'] = borough_stats['std_SAT'].round(2)

largest_std_dev = borough_stats.sort_values(by='std_SAT', ascending=False).head(1)
largest_std_dev = largest_std_dev.rename(columns={'Borough': 'borough'})

# ✅ Output (optional for debugging)
print("\n✅ Best Math Schools (Top 10):")
print(best_math_schools)

print("\n✅ Top 10 Schools by Total SAT:")
print(top_10_schools)

print("\n✅ Borough with Largest SAT Std Dev:")
print(largest_std_dev)
```
